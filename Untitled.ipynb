```python
def getDetails(car_details, carReg):
    for x in car_details:
        if carReg == x['VehicleRegistration']:
            return x
```

```python
carCollection = Cars.find()
car_details = []
for car in carCollection:
    car_details.append(car)
    
car_details
```

```python
for i, car in enumerate(cars):
  
    car_df = df.filter(df.CarRegistration == lit(car))
    
    days = getDays(car_df)
    
    vehicleTag = car.split(".csv")[0]
    
    details = getDetails(car_details, vehicleTag)
    
    
    for i,day in enumerate(days):
        
        temp = car_df.withColumn("YMD", date_format(df.Date, "yyyy MM dd"))
        temp = temp.filter(temp.YMD == day).drop("YMD")
        vehicle = Vehicle(vehicleTag, temp)
        vehicle.resample()
        time, velocity, lat, long = vehicle.getRegularTimestep()
        SP = zeroVP(time, velocity, lat, long,speedThresh=2,timeThresh=1000)
        
        ExportTrips(SP,time, velocity, lat, long, details, Trips, processed, Stops,Days)
        
        if i == 15:
            break
```
